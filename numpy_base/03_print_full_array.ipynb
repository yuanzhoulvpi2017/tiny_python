```python
import numpy as np 
```

```python
np.random.randint(0, 20, (10000,))
```

1. 可以看出上面，只是显示了几个数值，中间的很多数字都是被`...`给替代了。
2. 但是现在一个需求：不管这个向量多大，我都要把这个向量全部打印出来！应该怎么办？

## 解决办法
1. 使用`np.set_printoptions`函数就可以设置要打印数字的最大数量。可以把这个函数里面的`threshold`参数设置为`sys.maxsize`或者`np.inf`

```python
import sys
np.set_printoptions(threshold=sys.maxsize)# 其实np.set_printoptions(threshold=np.inf)也是可以的
# sys.maxsize
```

再来试一试，果然可以了

```python
np.random.randint(0, 20, (10000,))
```
