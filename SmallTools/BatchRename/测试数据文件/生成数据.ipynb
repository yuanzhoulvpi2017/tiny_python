```python
import pandas as pd
import os
os.makedirs(name='tif文件集合', exist_ok=True)
```

```python
rawdata = pd.read_excel("namen2o.xls")
rawdata
```

```python
def func(x):
    with open(file=f'tif文件集合/{x}', mode='w') as f:
        f.write(x)

_ = rawdata.pipe(
    lambda x: x['旧文件名称'].apply(
        lambda x: func(x)
    )
)
```
